# 3. Data Preparation

```python
df.drop(["num","t","sid","fname"], axis=1, inplace=True)
```

```python
sb.pairplot(df)
```

```python
df.head()
```

```python
df.corr()
```

```python
corr_matrix = df.corr()
corr_matrix["Tag.Mpa"].sort_values(ascending=False)
```

```python
X = df.drop(['Tag.Mpa'],axis=1)
```

```python
y = df['Tag.Mpa']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# 4. Modeling

```python
import sklearn.linear_model as sl

model = sl.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test.head(10)
```

```python
sb.scatterplot(x=y_test, y=y_pred)
```

```python
sb.displot(y_pred-y_test)
```

```python
import sklearn.tree as st

model1 = st.DecisionTreeRegressor(random_state=42)
```

```python
model1.fit(X_train, y_train)
```

```python
model1.score(X_test, y_test)
```

```python
y_pred1 = model1.predict(X_test)
```

```python
sb.scatterplot(x=y_test, y=y_pred1)
```

```python
sb.displot(y_pred1-y_test)
```

```python
import sklearn.ensemble as se

model2 = se.RandomForestRegressor(random_state=42)
```

```python
model2.fit(X_train, y_train)
```

```python
model2.score(X_test, y_test)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
sb.scatterplot(x=y_test, y=y_pred2)
```

```python
sb.displot(y_pred2-y_test)
```
